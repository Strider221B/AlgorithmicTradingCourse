# ADX - Average Directional Index

1. ADX is a way of measuring the strength of a Trend.
1. Values range from 0 to 100 and quantifies the strength of the Trend as below:
    1. 0-25: Absent or weak Trend
        1. Trend can change direction any point of time and we may end up losing money.
    1. 25-50: Strong Trend
    1. 50-75: Very Strong Trend
    1. 75-100: Extremly Strong Trend
1. ADX is non directional meaning that the ADX value makes no inference about the direction of the Trend but only about the strength of the Trend.
1. Calculation involves finding both positive and negative directional movements (by comparing successive highs and lows) and then calculating smoothed average of the difference of these.
