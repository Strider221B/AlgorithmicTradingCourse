```python
import datetime
import yfinance as yf
import pandas as pd
```

```python
stocks = ['AMZN', 'MSFT', 'FB', 'GOOG']
start = datetime.datetime.today() - datetime.timedelta(3650)
end = datetime.datetime.today()
cl_price = pd.DataFrame() # closing prices
ohl_cv = {}
```

```python
for ticker in stocks:
    cl_price[ticker] = yf.download(ticker, start, end)['Adj Close']
```

```python
cl_price.fillna(method='bfill', inplace=True)
```

```python
daily_return = cl_price.pct_change()
daily_return
```

```python
daily_return.mean()
```

```python
daily_return.std()
```

```python
daily_return.rolling(window=20).mean() # Simple moving average
```

```python
daily_return.rolling(window=20, min_periods=1).mean()
```

```python
daily_return.rolling(window=20).std()
```

### Exponential Moving Average

```python
daily_return.ewm(span=20, min_periods=20).mean()
```

```python
daily_return.ewm(span=20, min_periods=20).std() 
```
