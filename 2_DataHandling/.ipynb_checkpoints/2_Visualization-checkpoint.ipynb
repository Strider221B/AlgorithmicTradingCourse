```python
import datetime
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
stocks = ['AMZN', 'MSFT', 'FB', 'GOOG']
start = datetime.datetime.today() - datetime.timedelta(3650)
end = datetime.datetime.today()
cl_price = pd.DataFrame() # closing prices
ohl_cv = {}
```

```python
for ticker in stocks:
    cl_price[ticker] = yf.download(ticker, start, end)['Adj Close']
```

```python
cl_price.fillna(method='bfill', inplace=True)
daily_return = cl_price.pct_change()
```

```python
cl_price.plot(figsize=(10, 4));
```

```python
cp_standardized = (cl_price - cl_price.mean()) / cl_price.std()
cp_standardized.plot(figsize=(10, 4));
```

```python
cl_price.plot(subplots=True, figsize=(10, 10));
```

```python
plt.bar(daily_return.columns, daily_return.mean());
```
